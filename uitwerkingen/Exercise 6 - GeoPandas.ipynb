# Visualize CBS Wijkbuurtkaart with Geopandas

```python
import geopandas as gpd

import pandas as pd

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
```

### Get the data

```python
shapefile = r'../datasets/CBS Wijkbuurtkaart/gemeenten_2023_v1.shp'

gemeenten = gpd.read_file(shapefile)
```

```python
gemeenten.info()
```

```python
gemeenten = gemeenten[gemeenten['H2O']=='NEE'].copy()
```

```python
gemeenten.plot()
```

```python
gemeenten.info()
```

```python
gemeenten.iloc[:,:-3]
```

```python
gemeenten.plot(column='BEV_DICHTH', figsize=(10, 10), cmap='Spectral_r', legend=True)
```

```python
gemeenten[['GM_CODE','GM_NAAM','BEV_DICHTH']].nlargest(10, 'BEV_DICHTH')
```

```python
gemeenten[['GM_CODE','GM_NAAM','BEV_DICHTH']].nsmallest(10, 'BEV_DICHTH')
```

```python
gemeenten['MAN_VROUW'] = gemeenten['AANT_MAN'] / gemeenten['AANT_VROUW']
```

```python
gemeenten.plot(column='MAN_VROUW', figsize=(10, 10), cmap='Spectral_r', legend=True)
```

```python
gemeenten[['GM_CODE','GM_NAAM','MAN_VROUW']].nlargest(10, 'MAN_VROUW')
```

```python
gemeenten[['GM_CODE','GM_NAAM','MAN_VROUW']].nsmallest(10, 'MAN_VROUW')
```
