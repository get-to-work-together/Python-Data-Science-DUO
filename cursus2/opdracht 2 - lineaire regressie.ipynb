# Opdracht 2 - lineaire regressie

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2025)
```

```python
a = 2
b = 5
```

```python
x_data = np.random.uniform(0, 10, 100)
y_data = a * x_data + b
```

```python
spreiding = 3

ruis = np.random.randn(100) * spreiding

y_data = y_data + ruis
```

```python
plt.scatter(x_data, y_data)
plt.show()
```

```python
coefs = np.polyfit(x_data, y_data, 1)
coefs
```

```python
regressie = np.poly1d(coefs)
```

```python
x_model = np.linspace(0, 10, 100)
y_model = regressie(x_model)
```

```python
plt.scatter(x_data, y_data, color='darkgreen', alpha=0.4, s=100)
plt.plot(x_model, y_model, color='red')
plt.show()
```

## Metrics

```python
np.corrcoef(x_data, y_data)
```

```python
errors = y_data - regressie(x_data)
```

```python
mae = np.mean(np.abs(errors))
mae
```

```python
rmse = np.sqrt(np.mean(errors ** 2))
rmse
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_data, regressie(x_data))
```

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_data, regressie(x_data))
```

```python
from sklearn.metrics import r2_score
r2_score(y_data, regressie(x_data))
```
