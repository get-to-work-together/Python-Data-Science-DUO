# Diamonds price

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('diamonds')
```

```python
data.info()
```

```python
data.head()
```

```python
x_data = data['carat'].values
y_data = data['price'].values
```

```python
plt.scatter(x_data, y_data, s=20, alpha=0.5, color='green')
plt.show()
```

```python
coefs = np.polyfit(x_data, y_data, 2)

regressie = np.poly1d(coefs)

x_model = np.linspace(0, 5)
y_model = regressie(x_model)
```

```python
plt.scatter(x_data, y_data, s=20, alpha=0.5, color='green')
plt.plot(x_model, y_model, color='red')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_data, regressie(x_data))
```
