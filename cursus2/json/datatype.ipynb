# DATATYPE

```python
import pandas as pd
import json
```

```python
pd.set_option('display.max_rows', None)
```

```python
# filename = 'DATATYPE_V01.json'
filename = 'KENMERKWAARDEENUMERATIEWAARDE_V01.json'
```

```python
with open(filename) as f:
    s = json.load(f)
```

```python
df = pd.DataFrame(s['elements'])
df
```

```python
df_location_project = pd.concat([df, df['location'].str.split('::', expand=True)], axis=1)\
    .melt(id_vars=df.columns[:7])\
    .drop(['location', 'variable'], axis=1)\
    .dropna(subset='value')\
    .rename(columns={'value': 'location'})\
    .drop_duplicates()

df_location_project.head()
```

```python
df_location_project['location'].value_counts().sort_index()
```

### Verwijder rommel

```python
filter_queries = [
    'location.str.startswith("Speeltuin")',
    'location.str.startswith("Zandbak")',
    'location.str.startswith("Test")',
    'location in ["Elsbeth", "Engelien"]'
]

for q in filter_queries:
    df_location_project.drop(df_location_project.query(q).index, inplace=True)
```

### Opslaan als excel bestand

```python
df_location_project.to_excel('location_project.xlsx', index=False)
```
