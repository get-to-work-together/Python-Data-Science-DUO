```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.arange(0, 8*np.pi, 0.1)
a = 1
b = .2

for dt in np.arange(0, 2*np.pi, np.pi/2.0):

    x = a*np.cos(theta + dt)*np.exp(b*theta)
    y = a*np.sin(theta + dt)*np.exp(b*theta)

    dt = dt + np.pi/4.0

    x2 = a*np.cos(theta + dt)*np.exp(b*theta)
    y2 = a*np.sin(theta + dt)*np.exp(b*theta)

    xf = np.concatenate((x, x2[::-1]))
    yf = np.concatenate((y, y2[::-1]))

    p1 = plt.fill(xf, yf)

plt.show()
```

```python
x = np.linspace(0, 10, 40)
y = np.sin(x)/(x + 1)
```

```python
plt.plot(x, y)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].grid()
ax[0].axhline(0, color='black', linewidth=1)
ax[0].plot(x, y, linewidth=1, label='grafiek 1', marker='d')
ax[0].plot(x, y/2, linewidth=1, label='grafiek 2', marker='d')
ax[0].plot(x, y/3, linewidth=1, label='grafiek 3', marker='d')
ax[0].legend()
ax[0].set_title('Mijn voorbeeld')
ax[0].set_ylabel('y')

ax[1].set_xlabel('x')
ax[0].annotate('Kijk hier', 
               xy=(3.2, 0), 
               xytext=(4, 0.2), 
               arrowprops={'arrowstyle':'->'})

ax[1].scatter(x, y, color='C1', s=200, alpha=0.4)

import matplotlib as mpl
mpl.style.use('default')

plt.show()
```
