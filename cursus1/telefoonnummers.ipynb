```python
import pandas as pd
from collections import defaultdict

pd.set_option('display.max_rows', None)
```

```python
url = 'https://www.tuxx.nl/telefoneren/netnummers_nederland/'
```

```python
df = pd.read_html(url, header=0)[0]
df = df.drop('Unnamed: 1', axis='columns')
df['Netnummers'] = '0' + df['Netnummers'].astype('string')
df
```

```python
df = pd.concat([df['Netnummers'], df['Woonplaatsen'].str.split(',', expand=True)], axis='columns')
```

```python
df = df.melt(id_vars=['Netnummers']) \
       .drop('variable', axis=1) \
       .rename(columns={'value':'Woonplaats'}) \
       .dropna() \
       .sort_values('Netnummers') \
       .reset_index()
df['Woonplaats'] = df['Woonplaats'].str.strip()
```

```python
df_op_plaatsnaam = df.set_index('Woonplaats', drop=True) \
                     .drop('index', axis=1) \
                     .sort_index()
```
