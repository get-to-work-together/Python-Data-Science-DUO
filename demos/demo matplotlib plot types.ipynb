# Demo matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Line Plot

```python
x = np.linspace(0, 10, 101)
y = np.sin(x)/(x+1)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r.')
plt.show()
```

### Scatter Plot

```python
n = 200
x1 = np.random.randn(n)
x2 = np.random.randn(n)
```

```python
plt.scatter(x1, x2, c='darkviolet')
plt.show()
```

```python
sizes = np.random.uniform(10, 400, n)
colors = [['r', 'g', 'b'][i] for i in np.random.randint(0, 3, n)]
plt.scatter(x1, x2, marker='o', sizes=sizes, c=colors, alpha=0.4)
plt.show()
```

### Histogram

```python
n = 10000
data = np.random.randn(n)
```

```python
plt.hist(data, bins=20)
plt.show()
```

```python
bins = np.arange(-4, 4.1, 0.5)
plt.hist(data, bins=bins, edgecolor='darkgray', density=True, color='darkviolet')
plt.show()
```

### Boxplot

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data)
           
kwargs = dict(textcoords='offset pixels', xytext=(60, 0), verticalalignment='center', arrowprops={'arrowstyle':"->", 'color':'darkred'}, color='darkred')
plt.annotate('OUTLIERS', xy=(1.2, 3.3), **kwargs)
plt.annotate('MAXIMUM', xy=(1.2, 2.64), **kwargs)
plt.annotate('UPPER QUARTILE', xy=(1.2, 0.64), **kwargs)
plt.annotate('MEDIAN', xy=(1.2, 0.0), **kwargs)
plt.annotate('LOWER QUARTILE', xy=(1.2, -0.64), **kwargs)
plt.annotate('MINIMUM', xy=(1.2, -2.65), **kwargs)
plt.annotate('OUTLIERS', xy=(1.2, -3.3), **kwargs)

plt.gca().set(xlim=(0.7, 1.8))
plt.show()
```

### Violin Plot

```python
plt.violinplot(data)
plt.show()
```

```python
parts = plt.violinplot(data, showmeans=False, showmedians=False, showextrema=False)
for pc in parts['bodies']:
    pc.set_facecolor('#D43F3A')
    pc.set_edgecolor('black')
    pc.set_alpha(1)
plt.show()
```

### Bar chart

```python
d = {'Amsterdam': 934_927,
     'Rotterdam': 671_125,
     'Utrecht': 374_411,
     'Groningen': 243_833,
     'Den Haag': 565_701}

plt.bar(d.keys(), d.values())
plt.show()
```

```python
plt.barh(list(d.keys()), d.values(), color='cornflowerblue', edgecolor='darkblue', height=0.6, xerr=np.random.uniform(1_000, 200_000, len(d)), capsize=5, ecolor='darkred')
plt.show()
```

### Pie chart

```python
plt.pie(d.values(), labels = d.keys())
plt.show()
```

```python
plt.pie(d.values(), labels=d.keys(), explode=[0.10, 0, 0, 0, 0])
plt.show()
```

### Contour plot

```python
n = 100
x = np.linspace(-3.0, 3.0, n)
y = np.linspace(-2.0, 2.0, n)

X, Y = np.meshgrid(x, y)

Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
```

```python
plt.contour(X, Y, Z)
plt.show()
```

```python
plt.contourf(X, Y, Z, cmap=plt.cm.inferno)
plt.show()
```

### 3D

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
ax.view_init(elev=30., azim=-35, roll=0)

plt.show()
```

### Title

```python
fig, ax = plt.subplots(1, 1)

ax.plot(range(10), label='Line 1')
ax.plot(range(10)[::-1], label='Line 2', linestyle=(0, (3, 10)))

font_settings = {'fontsize': 24,
                 'fontweight': 'bold',
                 'color': 'darkred',
                 'verticalalignment': 'baseline',
                 'horizontalalignment': 'center'}

ax.set_title('Center Title', **font_settings)
ax.set_title('Left Title', loc='left')
ax.set_title('Right Title', loc='right')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

ax.legend()

plt.show()
```
