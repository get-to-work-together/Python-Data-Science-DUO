### Demo Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generate data

```python
definition = [
    {'mean': (0.2, 0.3), 'spread': (0.10, 0.10), 'n': 100},
    {'mean': (0.6, 0.7), 'spread': (0.10, 0.10), 'n': 100},
    {'mean': (0.8, 0.4), 'spread': (0.10, 0.10), 'n': 100},
]
```

```python
data_x = np.array([])
data_y = np.array([])

for cluster in definition:
    x = np.random.randn(cluster['n']) * cluster['spread'][0] + cluster['mean'][0]
    y = np.random.randn(cluster['n']) * cluster['spread'][1] + cluster['mean'][1]
    data_x = np.append(data_x, x)
    data_y = np.append(data_y, y)
```

```python
plt.scatter(data_x, data_y)
plt.show()
```
